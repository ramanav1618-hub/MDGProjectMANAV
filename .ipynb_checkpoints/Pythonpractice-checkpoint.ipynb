```python
dates=pd.date_range("20130101",periods=6)
column_names=[i for i in range(0,4)]
data=np.arange(1,25)
np.random.shuffle(data)
data=data.reshape(6,4)
df=pd.DataFrame(data,index=dates, columns=column_names)
print(df.iloc[0])
print()
print(df.iloc[0:1,0:4])
print()
print(df)
```

```python
dates=pd.date_range("20130101",periods=3)
column_names=[i for i in range(1,3)]
data=np.arange(1,7)
np.random.shuffle(data)
data=data.reshape(3,2)
df=pd.DataFrame(data,index=dates, columns=column_names)
print(df)
print()
print(df[(df[1]>=6)&(df[2]>=1)])
print()
```
